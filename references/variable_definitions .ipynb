# 變數定義說明文件 (Variable Definitions Documentation)

本文件詳細記錄本研究基於美國 CDC 青少年風險行為監測系統（YRBSS）資料集所篩選之核心研究變數、原始填答尺度，以及統計模型中衍生變項之操作定義。

---

## 一、 核心自變數：入門/門檻物質 (Core Independent Variables)

本研究鎖定三項青少年最常見的接觸物質作為核心預測變項。依據問卷設計，其原始數值皆為 **1.0 至 6.0 的分類級分**。

| :--- | :--- | :--- | :--- |
| `CurrentCigaretteUse` | **目前香菸使用頻率** | 過去 30 天內，受訪青少年抽菸的天數。 | **1**: 0天；**2**: 1-2天；**3**: 3-5天；**4**: 6-9天；**5**: 10-19天；**6**: 20天以上。 |
| `EverAlcoholUse` | **終生飲酒狀況** | 青少年終生累積的飲酒次數（不含宗教或家庭小酌）。 | **1**: 0次；**2**: 1-2次；**3**: 3-9次；**4**: 10-19次；**5**: 20-39次；**6**: 40次以上。 |

## 三、 專案衍生與推論統計變數 (Derived Variables)

為了發動微觀多維幾何分析與全局檢定，本專案在統計管線中透過編碼規則（Recoding Rules）建置了以下衍生變數：

### 1. 二元化涉入變項 (Binarized Substance Indicators)
* **`Cig_User`**：香菸二元涉入標籤（`1` = 過去30天有抽菸 級分 $\ge 2.0$；`0` = 無 級分 $= 1.0$）。
* **`Alc_User`**：酒精二元涉入標籤（`1` = 終生有飲酒 級分 $\ge 2.0$；`0` = 無 級分 $= 1.0$）。
* **`Mj_User`**：大麻二元涉入標籤（`1` = 過去30天有吸食 級分 $\ge 2.0$；`0` = 無 級分 $= 1.0$）。

### 2. 實驗分流與檢定變項 (Grouping & Multivariable Indicators)
* **`Substance_Group`**：**8 大互斥物質涉入群體變數**。將學生依據 `[Cig, Alc, Mj]` 的二元組合（如 `[0,0,0]` 到 `[1,1,1]`）精確分流至 G1 到 G8 八個互斥群體中，作為 ANOVA 變異數分析與 Tukey HSD 事後檢定的分組自變數。
* **`AnyHardDrug_Score`**：**綜合硬性毒品總體得分**（本研究全局檢定之核心應變數）。其操作定義為：
  
  $$\text{AnyHardDrug\_Score} = \max(\text{五項非法毒品原始級分}) - 1$$
  
  數值範圍為 **0 至 5 分**。`0` 代表完全沒碰過任何一項硬性毒品；分數越高代表該青少年涉入任何一種硬性毒品的最高頻率越嚴重。

### 3. 全局標準化自變數 (Standardized Regressors)
* **`Cig_scaled` / `Alc_scaled` / `Mj_scaled`**：全域標準化 OLS 多元線性迴歸專用變項。將自變數整數化後進行 Z-Score 轉換，使其轉化為「平均值 = 0，標準差 = 1」的無單位標準變數，用以直接比對並排序香菸、酒精與大麻對毒品的獨立引導權重（Standardized Beta）。

## III. Derived and Inferential Statistical Variables

To execute micro-level multidimensional geometric profiling and global testing, the following derived variables were constructed via the project's data pipeline based on specific recoding rules:

### 1. Binarized Substance Indicators
* **`Cig_User`**: Binary cigarette engagement indicator (`1` = Smoked in the past 30 days, raw score $\ge 2.0$; `0` = Abstinent, raw score $= 1.0$).
* **`Alc_User`**: Binary alcohol engagement indicator (`1` = Had alcohol over lifetime, raw score $\ge 2.0$; `0` = Abstinent, raw score $= 1.0$).
* **`Mj_User`**: Binary marijuana engagement indicator (`1` = Used marijuana in the past 30 days, raw score $\ge 2.0$; `0` = Abstinent, raw score $= 1.0$).

### 2. Experimental Stratification and Global Testing Indicators
* **`Substance_Group`**: **8 Mutually Exclusive Substance Engagement Groups Variable**. Stratifies individuals into groups G1 through G8 based on their binary combination of `[Cig, Alc, Mj]` (ranging from `[0,0,0]` to `[1,1,1]`). This serves as the key categorical factor for ANOVA and Tukey's HSD post-hoc multiple comparisons.
* **`AnyHardDrug_Score`**: **Composite Hard Drug Index/Score** (the primary dependent variable for global testing). It is operationally defined as:
  
  $$\text{AnyHardDrug\_Score} = \max(\text{Raw scores of the five hard drugs}) - 1$$
  
  The value spans from **0 to 5**. A score of `0` denotes complete abstinence from all five hard drugs; a higher score represents the peak severity/frequency of engagement in *any* of the tracked illicit substances.

### 3. Global Standardized Regressors
* **`Cig_scaled` / `Alc_scaled` / `Mj_scaled`**: Standard normal predictors used specifically for the global standardized OLS multiple linear regression. Constructed by converting the raw scales into integers and applying a Z-score transformation ("Mean = 0, Standard Deviation = 1"). This enables the direct comparison and ranking of independent standardized coefficients ($\beta$) to determine relative gateway weightings.
